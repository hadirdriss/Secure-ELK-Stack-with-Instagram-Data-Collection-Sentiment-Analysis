```python
import instaloader
import time
from pymongo import MongoClient
```

```python
#Défintion du paramètre de recherche; dans ce cas le décès du prèsident Jacques Chirac
```

```python
hashtag = "Décès président Jacques Chirac"
```

```python
# Fonction pour Récupèrer les posts instagram
```

```python
def chercher_posts_instagram(hashtag, nb = 5):
    """
    Fonction pour chercher les posts instagram
    """
    insta = instaloader.Instaloader(sleep = True, quiet = True)
    resultats = []

    try:
        tag = instaloader.Hashtag.from_name(insta.context, hashtag)
        for i, post in enumerate(tag.get_posts()):
            if i>= combien:
                break
            resultats.append({
            'utilisateur':post.owner_username,
            'texte': post.caption or '',
            'likes': post.likes,
            'date': str(post.date_utc),
            'lien': f"https://instagram.com/p/{post.shortcode}"
            })

            time.sleep(2)
        return resultats
    except:
        print("Problème de récupèration")
        return []
        
```

```python
posts = chercher_posts_instagram(hashtag,2)
```

```python
# Fonction pour enregistrer sur base de données MongoDB
```

```python
def save_to_mongo(posts):
    try:
        client=MongoClient('mongodb://localhossst:27017/')
        db=clent['instagram_db']
        collection=db['posts']
        compteur=0
        for post in posts:
            resultat=collection.update_one(
                {'id':post['id']},
                {'$set':post},
                upsert=True
            )
            if resultat.upserted_id:
                compteur += 1
        client.close()
        print(f"{compteur} posts sauvegardes")
        return True
    except Exception as e:
        print(f"Erreur MongoDB :{e}")
        return False 
```

```python
# Définir notre fonction MAIN
```

```python
def main():
    posts = chercher_posts_instagram(hashtag,2)
    if len(posts)>0:
        save_to_mongo(posts)
    print("Traitement terminé")
```

```python
main()
```
